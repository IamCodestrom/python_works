```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
#plot a simple function

price=[48000,54000,57000,49000,47000,45000]
year=[2015,2016,2017,2018,2019,2020]

plt.plot(year,price,marker='o')

plt.xlabel('Year')
plt.ylabel('price')
plt.title('Price Over the years')

plt.show()
```

```python
bats=pd.read_csv(r'D:\GdriveBackup\0.Projects\Mock Projects\DE\Python\datasets-session-23\sharma-kohli.csv')
bats

plt.plot(bats['index'],bats['V Kohli'],marker='o')
plt.plot(bats['index'],bats['RG Sharma'],marker='o')
```

```python
# titles for graph
plt.plot(bats['index'],bats['V Kohli'],marker='o')
plt.plot(bats['index'],bats['RG Sharma'],marker='o')

plt.title('Run Comparison between V Kohli and R Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.show()
```

```python
# colors(hex) and line(width and style) and marker(size)

plt.plot(bats['index'],bats['V Kohli'],marker='o',color='#D9F10F',linestyle='dashdot',linewidth=3,markersize=10)
plt.plot(bats['index'],bats['RG Sharma'],marker='D',color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Run Comparison between V Kohli and R Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.show()
```

```python
# legend -> location

plt.plot(bats['index'],bats['V Kohli'],marker='o',color='#D9F10F',linestyle='dashdot',linewidth=3,markersize=10,label='Virat')
plt.plot(bats['index'],bats['RG Sharma'],marker='D',color='#FC00D6',linestyle='dashdot',linewidth=2,label='Rohit')

plt.title('Run Comparison between V Kohli and R Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.legend(loc='lower center')
plt.show()
```

```python
# limiting axes

price=[48000,54000,57000,49000,47000,45000,4700000]
year=[2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price,marker='o')

plt.xlabel('Year')
plt.ylabel('price')
plt.title('Price Over the years')

plt.ylim(0,100000)
plt.xlim(2017,2020)
plt.show()
```

```python
# grid
plt.plot(bats['index'],bats['V Kohli'],marker='o',color='#D9F10F',linestyle='dashdot',linewidth=3,markersize=10,label='Virat')
plt.plot(bats['index'],bats['RG Sharma'],marker='D',color='#FC00D6',linestyle='dashdot',linewidth=2,label='Rohit')

plt.title('Run Comparison between V Kohli and R Sharma')
plt.xlabel('Years')
plt.ylabel('Runs')
plt.grid()
```

# Bar Chart

* Bivariate Analysis
* Categorical vs numercial data
* Use case- Agregate analysis of groups

```python
# simple bar chart

children=[10,20,30,10,30]
colors=['red','blue','green','yellow','pink']

plt.bar(colors,children,color=['red','blue','green','yellow','pink'])
```

```python
# horizontal bar chart

plt.barh(colors,children,color=['red','blue','green','yellow','pink'])
```

```python
bs=pd.read_csv(r'D:\GdriveBackup\0.Projects\Mock Projects\DE\Python\datasets-session-23\batsman_season_record.csv')
bs
```

```python
# multiple bar charts

plt.bar(np.arange(bs.shape[0])-0.2,bs['2015'],width=0.2,label='2015score')
plt.bar(np.arange(bs.shape[0]),bs['2016'],width=0.2,label='2016score')
plt.bar(np.arange(bs.shape[0])+0.2,bs['2017'],width=0.2,label='2017score')

plt.xticks(np.arange(bs.shape[0]),bs['batsman'])
# plt.bar(bs['batsman'],bs['2016'])

plt.legend()

plt.show()
```

```python
# problem
children=[10,20,30,10,30]
colors=['redredredred','blueblublublu','greengregregregre','yellowyelyelyelyel','pinkpinpinpin']

plt.bar(colors,children,color=['red','blue','green','yellow','pink'])
plt.xticks(rotation=80)
plt.show()
```

```python
# stacked bar chart
plt.bar(bs['batsman'],bs['2015'],label='2015score')
plt.bar(bs['batsman'],bs['2016'],label='2016score',bottom=bs['2015'])
plt.bar(bs['batsman'],bs['2017'],label='2017score',bottom=(bs['2016']+bs['2015']))

plt.legend(loc='upper left')
plt.show()
```
